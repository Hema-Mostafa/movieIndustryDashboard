# Dataset Mainuplation : 

```python
# remove duplicated from dataset:
movie_df.drop_duplicates(inplace=True)

# change date format
movie_df['release_date'] = pd.to_datetime(movie_df['release_date'])

# remove unWanted Columns from dataset
movie_df.drop(['budget_adj','revenue_adj','overview','imdb_id','homepage','tagline'],axis =1,inplace = True)

#remove rows with budget = 0  & revenue = 0 
indexes = movie_df[(movie_df['revenue'] == 0) | (movie_df['budget'] == 0)].index
movie_df.drop(index=indexes , inplace=True)

 
```
